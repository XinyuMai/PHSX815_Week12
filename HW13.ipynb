### Central Limit Theorem 

indicates that the normalized sum (i.e. the arithmetic mean) of random variables, each independently sampled from the same distribution, will asymptotically follow a Gaussian distribution. Try sampling a continuous variable from a probability distribution that is *not* a Gaussian 

- repeat this N times and take the average. Then repeat *that* again many times (so M "experiments" each with N samples) - what does this distribution of averages look like? As you increase N?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from Random import RandomDist

# default seed
seed = 5555
# class instance of our Random class using seed
random = RandomDist(seed)

def Beta_samples(a, b, Nsamples):
    return random.Beta(a, b, Nsamples)
```

```python
# start experiments

a = 1.5
b = 1.5
Nsamples = 100
Nexp = 100

values = []
for e in range(0, Nexp):
    samples = Beta_samples(a, b, Nsamples) 
    samples_avg = np.mean(samples)
    values.append(samples_avg)

# plotting
# best fit of data
plt.figure(figsize=(10,8))
(mu, sigma) = norm.fit(values) 
n, bins, patches = plt.hist(values, 50, density=True,  histtype='stepfilled', facecolor='g', alpha=0.5)                                                                                                                                                            
# plot formating options
y = norm.pdf(bins, mu, sigma)
plt.plot(bins, y, 'r--', linewidth=2)
plt.axvline(mu, color='r', linestyle='dashed',label ='$\\mu = %s$' % (mu))
plt.ylabel('Frequency', fontsize = 15)
plt.xlabel('sampled values ',fontsize = 15)
plt.grid(True)
plt.legend()
plt.show()
```
